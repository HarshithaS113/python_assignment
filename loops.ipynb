```python
## list of numbers square if postive, cube if negative ##
nums = [-3,-2,-1,0,1,2,3]
results = []
for i in nums:
    if i>0:
        results.append(i**2)
    elif i<0:
        results.append(i**3)
    else:
        results.append(0)
print(results)
```

```python
##marks = [30, 45, 60, 80, 25]

#if socre should be more than or equal to 40
# ["pass", "fail"]##

marks = [30, 45, 60, 80, 25]
results = []

for mark in marks:
    if mark >= 40:
        results.append("pass")
    else:
        results.append("fail")
print(results)

max_marks = max(marks)
print("maximum marks scored: ", max_marks)

min_marks = min(marks)
print("minimum marks scored: ", min_marks)

total_pass = results.count("pass")
print("total passed: ", total_pass)

total_fail = results.count("fail")
print("total failed: ", total_fail)
```

```python
##question 1
student_marks = int(input("Enter student's marks (0-100): "))

if student_marks < 0 or student_marks > 100:
    print("Invalid marks! Please enter a value between 0 and 100.")
else:
    print("Student's marks: ", student_marks)
    if student_marks >= 90:
        print("Grade A")
    elif student_marks >= 80:
        print("Grade B")
    elif student_marks >= 70:
        print("Grade C")
    elif student_marks >= 60:
        print("Grade D")
    else:
        print("Grade F")
```

```python
# A list of student scores
student_scores = [95, 29, 45, 76, 110, 68, -5]

print("Student Grading Results: \n")

for marks in student_scores:
    if marks < 0 or marks > 100:
        result = f"Score {marks} is Out of Range"
    elif marks >= 90:
        result = f"Score {marks}: Grade A"
    elif marks >= 80:
        result = f"Score {marks}: Grade B"
    elif marks >= 70:
        result = f"Score {marks}: Grade C"
    elif marks >= 60:
        result = f"Score {marks}: Grade D"
    else:
        result = f"Score {marks}: Grade F"
    
    print(result)
```

```python
#question 3 
a = int(input("enter a value: \n"))
b = int(input("enter a value: \n"))
c = int(input("enter a value: \n"))

print("triangle sides are:\n",a, b, c )

if a <= 0 or b <= 0 or c <= 0:
    print("Invalid triangle sides")
elif a + b <= c or a + c <= b or b + c <= a:
    print("Not a valid triangle")
else:
    if a == b == c: 
        print("it is a equilateral triangle \n",a , b, c)
    elif a == b or b==c or a==c:
        print("Isosceles triangle")
    else:
        print("Scalene triangle")
```

```python
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

print("Weight:", weight)
print("Height:", height)

BMI = weight / (height ** 2)
print("BMI:", BMI)

if BMI < 18.5:
    print("Underweight")
elif BMI < 25:
    print("Normal weight")
elif BMI < 30:
    print("Overweight")
else:
    print("Obese")
```
